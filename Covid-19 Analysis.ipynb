## Importing the dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv(r"https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv")
ds
```

## High Level Data Understanding

```python
# Number of rows & columns in the dataset

print(len(ds))           
print(len(ds.columns))   
```

```python
# Datatypes of columns

ds.dtypes
```

```python
ds.info()
```

```python
ds.describe()     # Statistical overview of only the numerical features
```

```python
ds.describe(include='all')       # Statistical overview of all the features
```

## Low Level Data Understanding

```python
# Finding count of unique values in location column.

ds['location'].nunique()
```

```python
# Finding the continent with the maximum frequency

max_freq = ds['continent'].value_counts().iloc[0:1]
max_freq
```

```python
# Finding the maximum & mean value in 'total_cases'.

total_cases_max = ds['total_cases'].max()
print(f"Maximum value in total cases: {total_cases_max}")

total_cases_mean = ds['total_cases'].mean()
print(f"Mean value in total cases: {total_cases_mean}")
```

```python
# Finding 25%,50% & 75% quartile value in 'total_deaths'

data = ds['total_deaths'].describe()
quartiles = data.iloc[4:7]
print(quartiles)
```

```python
# Finding which continent has minimum 'gdp_per_capita'.

gdp_per_capita = ds.groupby('continent').agg({'gdp_per_capita':'min'})
gdp_per_capita = gdp_per_capita.sort_values(by='gdp_per_capita',ascending=True)
print(gdp_per_capita)

min_value = gdp_per_capita.min()
print(f"The minimum gdp_per_capita is of the Africa continent:{min_value}")
```

```python
# Finding which continent has maximum 'human_development_index'.

human_development_index = ds.groupby('continent').agg({"human_development_index":"max"})
max_human_development_index = human_development_index.iloc[2,0:]
print(f" The maximum human development index is of Europe: {max_human_development_index}")
human_development_index
```

```python
# Filter the dataframe with only this columns

cols = ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']
ds = ds[cols]
ds
```

## Data Cleaning

```python
# Remove all duplicates observations

ds.duplicated()
```

```python
ds = ds.drop_duplicates()
```

```python
ds.duplicated().sum()
```

```python
# Finding the missing values in all columns

ds.isnull()
```

```python
ds.isnull().sum()
```

```python
ds = ds.dropna(subset=['continent'])
```

```python
ds.isnull().sum()
```

```python
# Fill all missing values with 0

ds = ds.fillna(0)
```

```python
ds.isnull().sum()
```

```python
ds
```

## Date time format :

```python
ds.dtypes
```

```python
# Converting date columns to datetime format

ds['date'] = pd.to_datetime(ds['date'], format='%d/%m/%y')
ds['month'] = ds['date'].dt.month
```

```python
# Create new column month after extracting month data from date column.

ds['month'] = ds['date'].dt.month
```

```python
ds
```

## Data Aggregation

```python
# Find max value in all columns using groupby function on 'continent' column
# Store the result in a new dataframe named 'df_groupby'.

x = ds.groupby('continent').agg({'location':'max','date':'max','total_cases':'max','total_deaths':'max','gdp_per_capita':'max','human_development_index':'max'})
x = x.reset_index()
df_groupby = x
df_groupby
```

## Feature Engineering

```python
# Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

df_groupby['total_deaths_to_total_cases'] = (df_groupby['total_deaths']/df_groupby['total_cases']).round(4)
```

```python
df_groupby
```

## Data Visualization

```python
# Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

plt.figure(figsize=(10,5))
sns.displot(df_groupby['gdp_per_capita'],kde=False) 
plt.xlabel("gdp per capita")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of 'total_cases' & 'gdp_per_capita'

plt.figure(figsize=(10,5))
df_groupby.plot.scatter(x='total_cases',y='gdp_per_capita',c='red',s=70)
plt.xlabel("Total Cases")
plt.ylabel("gdp per capita")
plt.tight_layout()
plt.show()
```

```python
# Pairplot on df_groupby dataset

plt.figure(figsize=(11,5))
sns.pairplot(data=df_groupby)
plt.tight_layout()
```

```python
# Bar plot of 'continent' column with 'total_cases'

plt.figure(figsize=(11,5))
sns.catplot(data=df_groupby, y='continent',x='total_cases',kind='bar',palette='Blues',legend=False,hue='continent')
plt.title("Total COVID-19 Cases by Continent")
plt.xlabel("Total Cases")
plt.ylabel("Continent")
plt.tight_layout()
plt.show()
```
